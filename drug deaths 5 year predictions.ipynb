```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load your dataset
# Assuming your data is in a DataFrame called 'df'

# Filter data for the particular city you want to predict
target_city = 'HARTFORD'
city_data = df[df['ResidenceCity'] == target_city]

# Data preprocessing
# Assuming 'Date' column represents the date of death
#city_data['Date'] = pd.to_datetime(city_data['Date'])
city_data.set_index('Date', inplace=True)

# Aggregate deaths per year
deaths_per_year = city_data.resample('Y').size()

# Visualize historical death trends
plt.figure(figsize=(10, 6))
plt.plot(deaths_per_year)
plt.title(f'Death Trends in {target_city} Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.show()

# Train the ARIMA model (Example: ARIMA(1,1,1))
model = ARIMA(deaths_per_year, order=(1, 1, 1))
trained_model = model.fit()

# Forecast death trends for the next 5 years
forecast_steps = 5
forecast = trained_model.forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=deaths_per_year.index[-1], periods=forecast_steps + 1, freq='Y')[1:]  # Generating forecast index




# Print the forecasted values
print(f"Forecasted death trends for the next {forecast_steps} years in {target_city}:")
print(forecast)
print(city_data.index.year.unique())
```
